```python
from gmpy2 import mpz, isqrt, is_prime, iroot

N_str = "62201672338679428490397681131579456733478193321837388980610095677399268773494625" + \
        "52709870902957883395760370472421354435369633128495863502078295716539396738183221" + \
        "59182133680794881103690005700730543266919010529237177075268417825077752933965287" + \
        "39045804210800100492546260275546725081590580479894723576877367054151343810079226" + \
        "0883517206472318827700593125450209"

N = mpz(N_str)

print(f"N are {len(N_str)} cifre")
print(f"N = {N}")
```

```python
def fermat(N):
    sqrt_n, exact = iroot(N, 2)
    
    if exact:
        print("N este patrat perfect")
        return sqrt_n, sqrt_n
    
    A = sqrt_n + 1
    
    print(f"sqrt(N) are {len(str(sqrt_n))} cifre")
    print(f"Incepem cautarea de la A = sqrt(N) + 1")
    
    for i in range(100000000):
        b_squared = A * A - N
        
        b, is_perfect = iroot(b_squared, 2)
        
        if is_perfect:
            p = A - b
            q = A + b
            
            print(f"\nFactorizare in {i+1} iteratii")
            print(f"A = {A}")
            print(f"b = {b}")
            return p, q
        
        A += 1
        
        if (i + 1) % 500000 == 0:
            print(f"Verificare {i+1} valori")
    
    print("Nu am putut gasi factorizarea")
    return None, None
```

```python
print("Metoda Fermat")

p, q = fermat(N)

if p and q:
    print("Factorizare cu succes")
else:
    print("\nNu s-au dat destule iteratii")
```

```python
print(f"\np = {p}")
print(f"\nq = {q}")
```

## Verificare

```python
print("Verificare ca numerele gasite sunt prime:")
print(f"p este prim: {is_prime(p)}")
print(f"q este prim: {is_prime(q)}")
```

```python
print(f"Verificare factorizare: p * q = N?")
print(f"Rezultat: {p * q == N}")
```

```python
# Calculare modul p - q
diff_pq = abs(p - q)
print(f"abs(p - q) = {diff_pq}")
```

```python
# Verificare conditie abs(p - q) < 2 * N^(1/4)
n_fourth_root = isqrt(isqrt(N))
bound = 2 * n_fourth_root
print(f"2 * N^(1/4) ~ {bound}")
print(f"Conditia abs(p - q) < 2 * N^(1/4) este satisfacuta: {diff_pq < bound}")
```
