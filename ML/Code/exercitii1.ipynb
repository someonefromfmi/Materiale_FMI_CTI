## 1. Fibonacci 
Scrieti o functie care primeste ca argument un numar natural `n` si afiseaza toti termenii din sirul lui Fibonacci mai mici decat `n`.

```python
def fib(n):
    if n == 1 or n == 2:
        return 1
    elif n == 0:
        return 0
    else:
        return fib(n-1) + fib(n-2)
    
print(fib(4))
    
```

```python
def fibList(n):
    i = 0
    while fib(i) < n:
        print(fib(i))
        i = i + 1
        
fibList(22)
```

## 2. Prime Number Check
Scrieti o functie care verifica daca un numar natural `n` este prim.

```python
import math

def isPrime(n):
    for i in range(2, int(math.sqrt(n))):
        if n % i == 0:
            return False
    return True

isPrime(23)
```
