## 1. Introducere teoretică

### Metoda KNN

KNN (K-Nearest Neighbors) este un algoritm de învățare automată supervizată care:

- **Clasifică** puncte noi bazându-se pe etichetele celor k vecini cei mai apropiați

- Utilizează **distanțe** între puncte (l1 - Manhattan, l2 - Euclideană)

- Este un algoritm **lazy** (nu învață un model explicit)


**Exemplu vizual:**
- NN classifier vs 5-NN classifier
- Zone colorate = regiune de decizie
- Metoda 1-NN poate crea 'insule' mici care duc la predicții incorecte
- Metoda 5-NN produce decizii mai smooth, dar poate avea zone gri de egalitate

### Setul de date MNIST
