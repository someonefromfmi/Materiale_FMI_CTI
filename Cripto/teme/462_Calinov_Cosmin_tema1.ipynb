```python
def xor_bytes(ciphertext, key):
    word_bytes = bytes.fromhex(ciphertext)
    return bytes([x ^ y for x, y in zip(word_bytes, key)])
```

```python
ciphertexts = [
    "2AC4B590FB9DBBDEE0146C93EADB95", # 1
    "2AC4B888F791B4C2EE086B9DE0DF8A", # 2
    "2AC8B689E69BBAC5E2057093E6D48A", # 3
    "29D5BE93FF93A8C5FB057093E6D48A", # 4
    "38DFB89AF781AFC5EE0A6596E0C994", # 5
    "38DEAE9FFA9DB0C3E60D679BE5D680", # 6
    "38C8A59AF791A8C5EE0A6D89FDD39A"  # 7
]
```

```python
word_list = open("wordlist.txt").read().split()
```

```python
for word in word_list:
    word_bytes = word.encode()
    candidate_key = xor_bytes(ciphertexts[0], word_bytes)
    valid_key_flag = True
    
    for enc_word in ciphertexts[1:]:
        dec_word = xor_bytes(enc_word, candidate_key)
        
        if dec_word.decode() not in word_list:
            valid_key_flag = False
            
    if valid_key_flag:
        print(f"Key: {candidate_key.hex()}")
        for i,c in enumerate(ciphertexts):
            m = xor_bytes(c, candidate_key).decode()
            print(f"Decrypted text: {i + 1}: {m}")
        break
```
