# Ex1

```python
# a
m = 10
r = 4

A = np.random.rand(m, r)

print(A)

print("Rang:\n")
np.linalg.matrix_rank(A)
```

```python
# b
b = np.array([
    A[:,0].copy() + A[:,1].copy() + A[:,2].copy(),
    A[:,1].copy() + A[:,2].copy() + A[:,3].copy(),
    A[:,1].copy() + A[:,2].copy() + A[:,3].copy() * 3,
    A[:,1].copy() + A[:,2].copy() + A[:,3].copy() * 2
])

A2 = np.hstack((A.copy(), b.T))

print(b)
print("\n")

print(A2)

print("Rang:")
np.linalg.matrix_rank(A2)
```

```python
# c
zgomot = np.random.normal(0, 0.2, b.T.shape)
print("\n")

matr_cu_zgomot = np.hstack((A.copy(), b.T + zgomot))

print(matr_cu_zgomot)
print("\n")

print("Rang:")
np.linalg.matrix_rank(matr_cu_zgomot)
```

```python
# d
U, S, V = np.linalg.svd(A2)

S
# Ultimele 4 valori sunt neglijabile. Da, rangul se poate deduce din matricea
# fara zgomot.
# Zgomotul adaugat mascheaza rangul real al unei matrici 
# (zgomotul perturba liniaritatea coloanelor din matrice, facand ca
# acestea sa aiba rang maxim)
# -> coloanele care erau inainte liniar dependendente au devenit
# independente datorita zgomotului introdus. Daca am aplica acum procedeul
# invers obtinerii combinatiilor liniare intre coloane, nu am mai obtine 0. 
```

```python
# a
babuin = img.imread('Baboon.bmp')
plt.imshow(babuin, cmap='gray')

print(babuin)
```

```python
U2, S2, V2  = np.linalg.svd(babuin.copy())

S2
```

```python
# b
def aprox_rang_k(u, s, v, k):
    U_k = u[:, :k]
    S_k = np.diag(s[:k])
    V_k = v[:k, :]
    
    img_aprox = np.dot(U_k, np.dot(S_k, V_k))
    
    return img_aprox

img_aprox50 = aprox_rang_k(U2, S2, V2, 50)
```

```python
# c
plt.imshow(img_aprox50, cmap='gray')
```

```python
# d

for k in [75, 255, 450]:
    new_aprox = aprox_rang_k(U2, S2, V2, k)
    plt.title('k = %d'% k)
    plt.imshow(new_aprox, cmap='gray')
    plt.show()
```

```python
data = pd.read_csv('iris.csv', header=None)

features = data.iloc[:, :4].apply(
    lambda x: (x - x.mean()) / x.std()
)
labels = data.iloc[:, 4]

pca = PCA(n_components=2)
components = pca.fit_transform(features)

categs = np.unique(labels)
plt.figure(figsize=(10,10))

for categ in categs:
    indices = labels == categ
    plt.scatter(components[indices, 0], components[indices, 1], label=categ)

plt.legend()
plt.grid()
plt.show()
```
